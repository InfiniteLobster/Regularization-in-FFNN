```python
import sys
from pathlib import Path
import numpy as np
```

```python
sys.path.insert(0, str(Path.cwd().parent / "src" / "AiB project"))
#
from model import FFNN
from activ import sigmoid, sigmoid_derivative, relu, relu_derivative #tanh, tanh_derivative
from init import init_zero, init_xavier_uniform
from loss import mse_loss, mse_loss_derivative
from update import standard
from train import train_GD, train_SGD
```

```python
input = np.array([[1, 3, 1, 4], [1, 7, 2, 5], [1, 7, 2, 5]])
output = np.array([[5.1, 0], [0, 2.1], [0, 3.1]])
print(input.shape)
print(output.shape)
```

```python
input_size = input.shape[1]
hidden_info = [3, 4, 3]
output_size = output.shape[1]
activ_func = [relu, relu, relu, sigmoid] 
activ_der = [relu_derivative, relu_derivative, relu_derivative, sigmoid_derivative]
activ_info = [activ_func, activ_der]
method_init = init_xavier_uniform
```

```python
model_1 = FFNN(input_size = input_size, hidden_info = hidden_info, output_size = output_size, activ_info = activ_info, method_init = method_init)
model_2 = FFNN(input_size = input_size, hidden_info = hidden_info, output_size = output_size, activ_info = activ_info, method_init = method_init)
model_3 = FFNN(input_size = input_size, hidden_info = hidden_info, output_size = output_size, activ_info = activ_info, method_init = method_init)
```

```python
out = model_1.forward(input)
print(out[-1])
```

```python
epochs = 100000
learning_rate = 0.01
for epoch in range(epochs):
    #forward
    out = model_1.forward(input)
    #loss
    loss = mse_loss(output, out[-1])
    if(epoch % 100 == 0):
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss}")
    #loss der
    loss_der = mse_loss_derivative(output, out[-1])
    #backward(update)
    model_1.backward(A = out, loss_derivative = loss_der, update_func = standard, learning_rate = learning_rate)
```

```python
model_2, loss_log = train_GD(model_2, input, output, update_func = standard, learning_rate = learning_rate, epochs = epochs)
print(loss_log)
```

```python
print(loss_log[-1])
print(loss_log[0])
```

```python
model_3, loss_log = train_SGD(model_3, input, output, update_func = standard, learning_rate = learning_rate, epochs = epochs)
print(loss_log)
```
